```python
frutas= {'banana': 1, 'laranja': 2, 'pera': 3, 'uva': 4}

for chave,valor in frutas.items():
  print(f"Para cada nome de {chave}, o valor dela é {valor}.")
  print("\n")

for chave in frutas.keys():
  print(chave)
  print(frutas[chave])
  print(f'Para cada nome de fruta {chave}.')
```
